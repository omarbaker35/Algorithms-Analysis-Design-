```python
from IPython.display import display, clear_output, Markdown
import ipywidgets as widgets
import pandas as pd
from datetime import datetime
import os
```

```python
SAVE_PATH = "inventory_data.csv"
```

```python
columns = ["פרופיל", "צבע", "כמות מקורית", "כמות נצרכה"]
df_inventory = pd.DataFrame(columns=columns)
```

```python
profile_images = {
    "30AP": "path_to_image",
    "30SC": "path_to_image",
    "30AP\\L": "path_to_image",
    "30TC": "path_to_image",
    "30TB": "path_to_image",
    "30MB": "path_to_image",
    "30DB": "path_to_image",
    "CC": "path_to_image",
    "30CC\\T\\B": "path_to_image",
    "30GG": "path_to_image",
    "30SLP": "path_to_image",
    "30LPSC": "path_to_image",
    "60AP": "path_to_image",
    "60SC": "path_to_image",
    "60AP\\L": "path_to_image",
    "60TC": "path_to_image",
    "60TB": "path_to_image",
    "60MB": "path_to_image",
    "60DB": "path_to_image",
    "60CC\\T\\B": "path_to_image",
    "60GG": "path_to_image",
    "60SLP": "path_to_image",
    "60LPSC": "path_to_image",
    "SLP63": "path_to_image"
}
```

```python
profiles_30 = ["30AP","30SC","30AP\\L","30TC","30TB","30MB","30DB","CC","30CC\\T\\B","30GG","30SLP","30LPSC"]
profiles_60 = ["60AP","60SC","60AP\\L","60TC","60TB","60MB","60DB","CC","60CC\\T\\B","60GG","60SLP","60LPSC"]
profiles_63 = profiles_60 + profiles_30 + ["SLP63"]

colors = ["לבן", "שחור", "כסף"]
actions = ["הוסף מלאי", "השתמש במלאי"]
```

```python
selected_type = ""
selected_color = ""
selected_action = ""
```

```python

def save_inventory():
    global df_inventory
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    df_inventory.to_csv(f"{timestamp}_inventory.csv", index=False)
    display(Markdown(f"✅ **DataFrame נשמר בהצלחה!** ({timestamp}_inventory.csv)"))
```

```python

def start_page():
    clear_output()
    start_btn = widgets.Button(description="התחל", button_style='success', layout=widgets.Layout(width='200px', height='50px'))
    display(start_btn)

    def on_start_clicked(b):
        choose_type_page()

    start_btn.on_click(on_start_clicked)
```

```python
def choose_type_page():
    clear_output()
    type_label = widgets.Label("בחר סוג: 30, 60 או 63")
    type_dropdown = widgets.Dropdown(options=["30","60","63"], description="", style={'description_width': 'initial'})
    next_btn = widgets.Button(description="הבא", button_style='info')
    back_btn = widgets.Button(description="חזור", button_style='warning')

    display(type_label, type_dropdown, widgets.HBox([back_btn, next_btn]))

    def on_next_clicked(b):
        global selected_type
        selected_type = type_dropdown.value
        choose_color_page()

    def on_back_clicked(b):
        start_page()

    next_btn.on_click(on_next_clicked)
    back_btn.on_click(on_back_clicked)
```

```python
def choose_color_page():
    clear_output()
    color_label = widgets.Label("בחר צבע:")
    color_dropdown = widgets.Dropdown(options=colors, description="", style={'description_width': 'initial'})
    next_btn = widgets.Button(description="הבא", button_style='info')
    back_btn = widgets.Button(description="חזור", button_style='warning')

    display(color_label, color_dropdown, widgets.HBox([back_btn, next_btn]))

    def on_next_clicked(b):
        global selected_color
        selected_color = color_dropdown.value
        choose_action_page()

    def on_back_clicked(b):
        choose_type_page()

    next_btn.on_click(on_next_clicked)
    back_btn.on_click(on_back_clicked)
```

```python

def choose_action_page():
    clear_output()
    action_label = widgets.Label("בחר פעולה:")
    action_dropdown = widgets.Dropdown(options=actions, description="", style={'description_width': 'initial'})
    next_btn = widgets.Button(description="הבא", button_style='info')
    back_btn = widgets.Button(description="חזור", button_style='warning')

    display(action_label, action_dropdown, widgets.HBox([back_btn, next_btn]))

    def on_next_clicked(b):
        global selected_action
        selected_action = action_dropdown.value
        enter_values_page()

    def on_back_clicked(b):
        choose_color_page()

    next_btn.on_click(on_next_clicked)
    back_btn.on_click(on_back_clicked)
```

```python
def enter_values_page():
    clear_output()

    if selected_type == "30":
        profiles = profiles_30
    elif selected_type == "60":
        profiles = profiles_60
    else:
        profiles = profiles_63

    qty_widgets = {}
    display(Markdown(f"**בחר כמויות עבור כל פרופיל ({selected_color})**"))

    for profile in profiles:
        label = widgets.Label(profile)
        qty_box = widgets.IntText(value=0, description='כמות:', layout=widgets.Layout(width='150px'))
        display(widgets.HBox([label, qty_box]))
        qty_widgets[profile] = qty_box

    save_btn = widgets.Button(description="חשב ושמור", button_style='success')
    back_btn = widgets.Button(description="חזור", button_style='warning')

    display(widgets.HBox([back_btn, save_btn]))

    def on_save_clicked(b):
        global df_inventory
        for profile, widget in qty_widgets.items():
            qty = widget.value
            if selected_action == "הוסף מלאי":
                if ((df_inventory['פרופיל']==profile) & (df_inventory['צבע']==selected_color)).any():
                    df_inventory.loc[(df_inventory['פרופיל']==profile) & (df_inventory['צבע']==selected_color), 'כמות מקורית'] += qty
                else:
                    df_inventory = pd.concat([df_inventory, pd.DataFrame([[profile, selected_color, qty, 0]], columns=columns)], ignore_index=True)
            else:
                if ((df_inventory['פרופיל']==profile) & (df_inventory['צבע']==selected_color)).any():
                    df_inventory.loc[(df_inventory['פרופיל']==profile) & (df_inventory['צבע']==selected_color), 'כמות נצרכה'] += qty
                    df_inventory.loc[(df_inventory['פרופיל']==profile) & (df_inventory['צבע']==selected_color), 'כמות מקורית'] -= qty
        display(df_inventory)
        save_inventory()

    def on_back_clicked(b):
        choose_action_page()

    save_btn.on_click(on_save_clicked)
    back_btn.on_click(on_back_clicked)
```
